# Wave Array

## Difficulty
Medium

## Description
Given a sorted array `arr[]` of integers, rearrange the array into a wave-like form **in place**.

The wave condition is:
`arr[0] >= arr[1] <= arr[2] >= arr[3] <= arr[4] ...`

If there are multiple valid wave arrangements, return the **lexicographically smallest** one.

The array is already sorted in ascending order and must be modified in place without returning a new array.

## Examples

**Example 1**
- Input: `arr[] = [1, 2, 3, 4, 5]`
- Output: `[2, 1, 4, 3, 5]`
- Explanation: The array is rearranged to satisfy the wave condition.

**Example 2**
- Input: `arr[] = [2, 4, 7, 8, 9, 10]`
- Output: `[4, 2, 8, 7, 10, 9]`
- Explanation: The array follows the wave pattern.

**Example 3**
- Input: `arr[] = [1]`
- Output: `[1]`

## Constraints
- `1 ≤ arr.size ≤ 10^6`
- `0 ≤ arr[i] ≤ 10^9`


```python
class Solution:
    def sortInWave(self, arr):
        for i in range(len(arr)):
            if i%2 == 0 and i < len(arr)-1:
                arr[i],arr[i+1] = arr[i+1],arr[i] 
                
```

# Equilibrium Point

## Difficulty
Easy

## Description
Given an array of integers `arr[]`, find the **first equilibrium point** in the array.

An equilibrium point is an index (0-based) such that the sum of all elements before that index is equal to the sum of all elements after it.

Return `-1` if no such index exists.

## Examples

**Example 1**
- Input: `arr[] = [1, 2, 0, 3]`
- Output: `2`
- Explanation: Sum of elements before index 2 is `1 + 2 = 3`, and sum of elements after index 2 is `3`.

**Example 2**
- Input: `arr[] = [1, 1, 1, 1]`
- Output: `-1`
- Explanation: No equilibrium index exists.

**Example 3**
- Input: `arr[] = [-7, 1, 5, 2, -4, 3, 0]`
- Output: `3`
- Explanation: Sum of elements before index 3 is `-7 + 1 + 5 = -1`, and sum of elements after index 3 is `-4 + 3 + 0 = -1`.

## Constraints
- `3 ≤ arr.size() ≤ 10^5`
- `-10^4 ≤ arr[i] ≤ 10^4`


```python
# User function Template for python3

class Solution:
    #Function to find equilibrium point in the array.
    def findEquilibrium(self, arr):
        
        left_sum = 0
        right_sum = sum(arr)
        
        for i in range(len(arr)):
            
            right_sum = right_sum - arr[i]
            
            if left_sum == right_sum:
                return i
            
            left_sum += arr[i]
            
        return -1
        
        
```

# Missing in Array

## Difficulty
Easy

## Description
You are given an array `arr[]` of size `n - 1` containing distinct integers in the range from `1` to `n` (inclusive).  
The array represents a permutation of numbers from `1` to `n` with **one element missing**.

Your task is to find and return the missing element.

## Examples

**Example 1**
- Input: `arr[] = [1, 2, 3, 5]`
- Output: `4`
- Explanation: All numbers from 1 to 5 are present except 4.

**Example 2**
- Input: `arr[] = [8, 2, 4, 5, 3, 7, 1]`
- Output: `6`
- Explanation: All numbers from 1 to 8 are present except 6.

**Example 3**
- Input: `arr[] = [1]`
- Output: `2`
- Explanation: Only 1 is present, so the missing number is 2.

## Constraints
- `1 ≤ arr.size() ≤ 10^6`
- `1 ≤ arr[i] ≤ arr.size() + 1`


```python
class Solution:
    def missingNum(self, arr):
        
        n = len(arr)+1
        
        expected_sum = n * (n + 1) // 2
        actual_sum = sum(arr)
        
        return expected_sum - actual_sum
```

# 1614. Maximum Nesting Depth of the Parentheses

## Difficulty

## Description
Given a valid parentheses string `s`, return the nesting depth of `s`.

The nesting depth is defined as the maximum number of nested parentheses at any point in the string.

## Examples

**Example 1**
- Input: `s = "(1+(2*3)+((8)/4))+1"`
- Output: `3`
- Explanation: Digit `8` is inside 3 nested parentheses.

**Example 2**
- Input: `s = "(1)+((2))+(((3)))"`
- Output: `3`
- Explanation: Digit `3` is inside 3 nested parentheses.

**Example 3**
- Input: `s = "()(())((()()))"`
- Output: `3`

## Constraints
- `1 ≤ s.length ≤ 100`
- `s` consists of digits `0-9` and characters `'+'`, `'-'`, `'*'`, `'/'`, `'('`, and `')'`
- The parentheses expression `s` is guaranteed to be valid


```python
class Solution:
    def maxDepth(self, s: str) -> int:
        max_dpt = 0
        count = 0
        for i in s:
            if i == '(':
                count += 1
            elif i == ')':
                max_dpt = max(max_dpt,count)
                count -= 1
            else:
                pass
        return max_dpt

            
```

# 13. Roman to Integer

## Difficulty
Easy

## Description
Roman numerals are represented by seven different symbols with the following values:

- `I` → 1  
- `V` → 5  
- `X` → 10  
- `L` → 50  
- `C` → 100  
- `D` → 500  
- `M` → 1000  

Roman numerals are generally written from largest to smallest values from left to right. However, in certain cases, subtraction is used:
- `I` before `V` or `X` → 4 or 9  
- `X` before `L` or `C` → 40 or 90  
- `C` before `D` or `M` → 400 or 900  

Given a Roman numeral string `s`, convert it to an integer.

## Examples

**Example 1**
- Input: `s = "III"`
- Output: `3`

**Example 2**
- Input: `s = "LVIII"`
- Output: `58`
- Explanation: `L = 50`, `V = 5`, `III = 3`

**Example 3**
- Input: `s = "MCMXCIV"`
- Output: `1994`
- Explanation: `M = 1000`, `CM = 900`, `XC = 90`, `IV = 4`

## Constraints
- `1 ≤ s.length ≤ 15`
- `s` contains only the characters `'I'`, `'V'`, `'X'`, `'L'`, `'C'`, `'D'`, `'M'`
- `s` is guaranteed to be a valid Roman numeral in the range `[1, 3999]`


```python
class Solution:
    def romanToInt(self, s: str) -> int:
        rom_int_map = {'I': 1, 'V': 5, 'X': 10, 'L': 50, 'C': 100, 'D': 500, 'M': 1000}
        m=len(s)
        out = 0
        for i in range(m):
            if i+1 <= m-1:
                if rom_int_map[s[i]] < rom_int_map[s[i+1]]:
                    out -=  rom_int_map[s[i]] 
                    continue
                
            out +=  rom_int_map[s[i]] 
        
        return out
            
```

# 205. Isomorphic Strings

## Difficulty
Easy

## Description
Given two strings `s` and `t`, determine whether they are isomorphic.

Two strings are isomorphic if the characters in `s` can be replaced to get `t`, such that:
- All occurrences of a character must be replaced with the same character.
- No two characters may map to the same character.
- A character may map to itself.
- The order of characters must be preserved.

## Examples

**Example 1**
- Input: `s = "egg"`, `t = "add"`
- Output: `true`
- Explanation:  
  `'e' → 'a'`, `'g' → 'd'`

**Example 2**
- Input: `s = "foo"`, `t = "bar"`
- Output: `false`
- Explanation:  
  `'o'` would need to map to both `'a'` and `'r'`.

**Example 3**
- Input: `s = "paper"`, `t = "title"`
- Output: `true`

## Constraints
- `1 ≤ s.length ≤ 5 × 10^4`
- `t.length == s.length`
- `s` and `t` consist of valid ASCII characters


```python
class Solution:
    def isIsomorphic(self, s: str, t: str) -> bool:
        
        s_t_map = {}
        t_s_map = {}

        for i in range(len(s)):
            if s[i] in s_t_map.keys() and s_t_map[s[i]] != t[i]:
                return False
            if t[i] in t_s_map.keys() and t_s_map[t[i]] != s[i]:
                return False
            
            s_t_map[s[i]] = t[i]
            t_s_map[t[i]] = s[i]
        
        return True
```

# 242. Valid Anagram

## Difficulty
Easy

## Description
Given two strings `s` and `t`, determine whether `t` is an anagram of `s`.

An anagram is formed by rearranging the letters of a string using all the original characters exactly once.

## Examples

**Example 1**
- Input: `s = "anagram"`, `t = "nagaram"`
- Output: `true`

**Example 2**
- Input: `s = "rat"`, `t = "car"`
- Output: `false`

## Constraints
- `1 ≤ s.length, t.length ≤ 5 × 10^4`
- `s` and `t` consist of lowercase English letters

## Follow-up
- What if the inputs contain Unicode characters? How would you adapt your solution to handle this case?


```python
class Solution:
    def isAnagram(self, s: str, t: str) -> bool:
        if len(s) != len(t):
            return False
        s_map = {}
        t_map = {}
        
        for i in s:
            if i in s_map.keys():
                s_map[i] += 1
            else:
                s_map[i] = 1
                
        for i in t:
            if i in t_map.keys():
                t_map[i] += 1
            else:
                t_map[i] = 1
        
        for k in s_map.keys():
            if k not in t_map.keys():
                return False
            if s_map[k] != t_map[k]:
                return False
        return True
```

# Minimum Swaps and K Together

## Difficulty
Medium

## Description
Given an array `arr[]` and an integer `k`, you can perform a swap operation any number of times by choosing any two indices `i` and `j` (`i < j`) and swapping `arr[i]` and `arr[j]`.

The task is to find the **minimum number of swaps** required to bring all the elements **less than or equal to `k` together**, i.e., make them form a contiguous subarray.

## Examples

**Example 1**
- Input: `arr[] = [2, 1, 5, 6, 3]`, `k = 3`
- Output: `1`
- Explanation: Swap elements `5` and `3` to group `{2, 1, 3}` together.

**Example 2**
- Input: `arr[] = [2, 7, 9, 5, 8, 7, 4]`, `k = 6`
- Output: `2`
- Explanation: Two swaps are needed to bring elements `{2, 5, 4}` together.

**Example 3**
- Input: `arr[] = [2, 4, 5, 3, 6, 1, 8]`, `k = 6`
- Output: `0`
- Explanation: All elements less than or equal to `k` are already contiguous.

## Constraints
- `1 ≤ arr.size() ≤ 10^6`
- `1 ≤ arr[i] ≤ 10^6`
- `1 ≤ k ≤ 10^6`


```python
#User function Template for python3

class Solution:
        
    def minSwap (self,arr, k) : 
        
        count = 0
        
        for i in arr:
            if i <= k:
                count += 1
                
        bad = 0
        for i in range(count):
            if arr[i] > k:
                bad += 1
                
        
        min_swap = bad
        
        for i in range(count,len(arr)):
            
            if arr[i-count] > k:
                bad -= 1
            
            if arr[i] > k:
                bad += 1
            
            
            min_swap = min(min_swap,bad)
        
        
        return min_swap 
```

# 189. Rotate Array

## Difficulty
Medium

## Description
Given an integer array `nums`, rotate the array to the right by `k` steps, where `k` is non-negative.

The rotation should rearrange the elements so that each element is shifted to the right by `k` positions.

## Examples

**Example 1**
- Input: `nums = [1,2,3,4,5,6,7]`, `k = 3`
- Output: `[5,6,7,1,2,3,4]`
- Explanation:  
  After 1 rotation: `[7,1,2,3,4,5,6]`  
  After 2 rotations: `[6,7,1,2,3,4,5]`  
  After 3 rotations: `[5,6,7,1,2,3,4]`

**Example 2**
- Input: `nums = [-1,-100,3,99]`, `k = 2`
- Output: `[3,99,-1,-100]`
- Explanation:  
  After 1 rotation: `[99,-1,-100,3]`  
  After 2 rotations: `[3,99,-1,-100]`

## Constraints
- `1 ≤ nums.length ≤ 10^5`
- `-2^31 ≤ nums[i] ≤ 2^31 - 1`
- `0 ≤ k ≤ 10^5`

## Follow-up
- Can you solve the problem in-place with `O(1)` extra space?
- Can you think of multiple approaches to solve this problem?


```python
class Solution:
    def rotate(self, nums: List[int], k: int) -> None:
        """
        Do not return anything, modify nums in-place instead.
        """
        n = len(nums)
        k = k%n
        nums[:] = nums[n-k:] + nums[:n-k]
```

# 567. Permutation in String

## Difficulty
Medium

## Description
Given two strings `s1` and `s2`, determine whether `s2` contains a permutation of `s1`.

In other words, return `true` if any permutation of `s1` appears as a contiguous substring of `s2`, otherwise return `false`.

## Examples

**Example 1**
- Input: `s1 = "ab"`, `s2 = "eidbaooo"`
- Output: `true`
- Explanation: `s2` contains a permutation of `s1` (`"ba"`).

**Example 2**
- Input: `s1 = "ab"`, `s2 = "eidboaoo"`
- Output: `false`

## Constraints
- `1 ≤ s1.length, s2.length ≤ 10^4`
- `s1` and `s2` consist of lowercase English letters


```python
class Solution:
    def checkInclusion(self, s1: str, s2: str) -> bool:
        s1_hash = {}
        s2_hash = {}

        per_found = False

        for i in s1:
            if i in s1_hash.keys():
                s1_hash[i] = s1_hash[i] + 1
            else:
                s1_hash[i] = 1
            
        first = 0
        last = len(s1)

        while last <= len(s2):
            for i in s2[first:last]:
                if i in s2_hash.keys():
                    s2_hash[i] = s2_hash[i] + 1
                else:
                    s2_hash[i] = 1
            
            for key,value in s1_hash.items():
                if key in s2_hash.keys() and s2_hash[key] == value:
                    per_found = True
                    continue
                else:
                    per_found = False
                    break
            
            if per_found:
                return per_found

            first += 1
            last += 1
            s2_hash = {}



        return per_found
```

# 229. Majority Element II

## Difficulty
Medium

## Description
Given an integer array `nums` of size `n`, find all elements that appear more than ⌊ n / 3 ⌋ times.

Return the elements in any order.

## Examples

**Example 1**
- Input: `nums = [3,2,3]`
- Output: `[3]`

**Example 2**
- Input: `nums = [1]`
- Output: `[1]`

**Example 3**
- Input: `nums = [1,2]`
- Output: `[1,2]`

## Constraints
- `1 ≤ nums.length ≤ 5 × 10^4`
- `-10^9 ≤ nums[i] ≤ 10^9`

## Follow-up
- Can you solve the problem in linear time and `O(1)` extra space?


```python
class Solution:
    def majorityElement(self, nums: List[int]) -> List[int]:
        n = len(nums)//3

        count_map = {}

        for i in nums:
            if i in count_map.keys():
                count_map[i] = count_map[i] + 1
            
            else:
                count_map[i] = 1
        
        out = []
        for key,value in count_map.items():
            if value > n:
                out.append(key)

        return out
```

# 287. Find the Duplicate Number

## Difficulty
Medium

## Description
Given an array of integers `nums` containing `n + 1` integers where each integer is in the range `[1, n]` inclusive, there is exactly one integer that appears more than once.

Return the duplicate number.

You must solve the problem **without modifying** the array and using **only constant extra space**.

## Examples

**Example 1**
- Input: `nums = [1,3,4,2,2]`
- Output: `2`

**Example 2**
- Input: `nums = [3,1,3,4,2]`
- Output: `3`

**Example 3**
- Input: `nums = [3,3,3,3,3]`
- Output: `3`

## Constraints
- `1 ≤ n ≤ 10^5`
- `nums.length == n + 1`
- `1 ≤ nums[i] ≤ n`
- Exactly one integer appears two or more times

## Follow-up
- How can we prove that at least one duplicate number must exist in `nums`?
- Can you solve the problem in linear time complexity?


```python
class Solution:
    def findDuplicate(self, nums: List[int]) -> int:
        slow = 0
        fast = 0
        while True:
            slow = nums[slow]
            fast = nums[nums[fast]]

            if slow == fast:
                break

        slow = 0

        while slow != fast:
            slow = nums[slow]
            fast = nums[fast]

        return slow
```

# 442. Find All Duplicates in an Array

## Difficulty
Medium

## Description
Given an integer array `nums` of length `n` where all integers are in the range `[1, n]` and each integer appears at most twice, return an array of all the integers that appear exactly twice.

You must write an algorithm that runs in `O(n)` time and uses only constant extra space, excluding the space required for the output.

## Examples

**Example 1**
- Input: `nums = [4,3,2,7,8,2,3,1]`
- Output: `[2,3]`

**Example 2**
- Input: `nums = [1,1,2]`
- Output: `[1]`

**Example 3**
- Input: `nums = [1]`
- Output: `[]`

## Constraints
- `n == nums.length`
- `1 ≤ n ≤ 10^5`
- `1 ≤ nums[i] ≤ n`
- Each element in `nums` appears once or twice


```python
class Solution:
    def findDuplicates(self, nums: List[int]) -> List[int]:
        out = []


        for n in nums:
            index = abs(n)-1
            if nums[index] < 0:
                out.append(abs(n))

            else:
                nums[index] = - nums[index]


        return out
```

# 152. Maximum Product Subarray

## Difficulty
Medium

## Description
Given an integer array `nums`, find a contiguous subarray that has the largest product and return the product.

The product of an array with a single element is the value of that element.

The result is guaranteed to fit in a 32-bit integer.

## Examples

**Example 1**
- Input: `nums = [2,3,-2,4]`
- Output: `6`
- Explanation: The subarray `[2,3]` has the largest product.

**Example 2**
- Input: `nums = [-2,0,-1]`
- Output: `0`
- Explanation: The result cannot be 2 because `[-2,-1]` is not a contiguous subarray.

## Constraints
- `1 ≤ nums.length ≤ 2 × 10^4`
- `-10 ≤ nums[i] ≤ 10`
- The product of any subarray fits in a 32-bit integer


```python
class Solution:
    def maxProduct(self, nums: List[int]) -> int:
        prefix = 1
        sufix = 1
        max_prod = float('-inf')
        n = len(nums)

        for i in range(len(nums)):
            if prefix == 0:
                prefix = 1
            if sufix == 0:
                sufix = 1
            prefix *= nums[i]
            sufix *= nums[n-1-i]
            max_prod = max(max_prod,max(prefix,sufix))

        return max_prod
```
