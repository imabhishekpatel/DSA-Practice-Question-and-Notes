# Reverse String

## Difficulty
Easy

## Description
Write a function that reverses a string. The input string is given as an array of characters `s`.

You must reverse the string **in-place** using **O(1)** extra memory.

## Examples

**Example 1**
- Input: `s = ["h","e","l","l","o"]`
- Output: `["o","l","l","e","h"]`

**Example 2**
- Input: `s = ["H","a","n","n","a","h"]`
- Output: `["h","a","n","n","a","H"]`

## Constraints
- `1 ≤ s.length ≤ 10^5`
- `s[i]` is a printable ASCII character


```python
class Solution:
    def reverseString(self, s: List[str]) -> None:
        """
        Do not return anything, modify s in-place instead.
        """

        def rev_str(left,right):
            if left>=right:
                return
            s[left],s[right] = s[right],s[left]

            rev_str(left+1,right-1)
        
        rev_str(0,len(s)-1)


        
```

# Fibonacci Number

## Difficulty
Easy

## Description
The Fibonacci numbers form a sequence such that each number is the sum of the two preceding ones, starting from 0 and 1.

The sequence is defined as:
- `F(0) = 0`
- `F(1) = 1`
- `F(n) = F(n - 1) + F(n - 2)` for `n > 1`

Given an integer `n`, calculate `F(n)`.

## Examples

**Example 1**
- Input: `n = 2`
- Output: `1`
- Explanation: `F(2) = F(1) + F(0) = 1 + 0 = 1`

**Example 2**
- Input: `n = 3`
- Output: `2`
- Explanation: `F(3) = F(2) + F(1) = 1 + 1 = 2`

**Example 3**
- Input: `n = 4`
- Output: `3`
- Explanation: `F(4) = F(3) + F(2) = 2 + 1 = 3`

## Constraints
- `0 ≤ n ≤ 30`


```python
class Solution:
    def fib(self, n: int) -> int:
        # base case

        if n == 0:
            return 0
        if n == 1:
            return 1

        return self.fib(n-1) + self.fib(n-2)
            

        
```

# Power of Two

## Difficulty
Easy

## Description
Given an integer `n`, return `true` if it is a power of two. Otherwise, return `false`.

An integer `n` is a power of two if there exists an integer `x` such that:
`n == 2^x`.

## Examples

**Example 1**
- Input: `n = 1`
- Output: `true`
- Explanation: `2^0 = 1`

**Example 2**
- Input: `n = 16`
- Output: `true`
- Explanation: `2^4 = 16`

**Example 3**
- Input: `n = 3`
- Output: `false`

## Constraints
- `-2^31 ≤ n ≤ 2^31 - 1`

## Follow-up
- Could you solve it without loops or recursion?


```python
class Solution:
    def isPowerOfTwo(self, n: int) -> bool:
        
        def powTwo(x):

            if x == 1:
                return True
            if x%2 != 0:
                return False
            if x == 0:
                return False
            
            
            return powTwo(x/2)
        
        return powTwo(n)
```

# Pow(x, n)

## Difficulty
Medium

## Description
Implement `pow(x, n)`, which calculates `x` raised to the power `n` (i.e., `xⁿ`).

The value of `n` can be positive, negative, or zero.

## Examples

**Example 1**
- Input: `x = 2.00000, n = 10`
- Output: `1024.00000`

**Example 2**
- Input: `x = 2.10000, n = 3`
- Output: `9.26100`

**Example 3**
- Input: `x = 2.00000, n = -2`
- Output: `0.25000`
- Explanation: `2⁻² = 1 / 2² = 1 / 4 = 0.25`

## Constraints
- `-100.0 < x < 100.0`
- `-2^31 ≤ n ≤ 2^31 - 1`
- `n` is an integer
- Either `x` is not zero or `n > 0`
- `-10^4 ≤ xⁿ ≤ 10^4`


```python
class Solution:
    def myPow(self, x: float, n: int) -> float:
        def power(x, n):
            if n == 0:
                return 1
            
            half = power(x, n // 2)
            
            if n % 2 == 0:
                return half * half
            else:
                return half * half * x
        
        if n < 0:
            x = 1 / x
            n = -n
        
        return power(x, n)
```

# Generate Parentheses

## Difficulty
Medium

## Description
Given `n` pairs of parentheses, generate all combinations of **well-formed parentheses**.

A combination is well-formed if every opening parenthesis has a corresponding closing parenthesis and the parentheses are properly nested.

## Examples

**Example 1**
- Input: `n = 3`
- Output: `["((()))","(()())","(())()","()(())","()()()"]`

**Example 2**
- Input: `n = 1`
- Output: `["()"]`

## Constraints
- `1 ≤ n ≤ 8`


```python
class Solution:
    def generateParenthesis(self, n: int) -> List[str]:
        ans = []

        def addParen(open,close,ans,curr):

            # base case
            if open == 0 and close == 0:
                ans.append(curr)
                return 

            if open != 0:
                addParen(open-1,close,ans,curr+'(')
            if close > open:
                addParen(open,close-1,ans,curr+')')

        addParen(n,n,ans,"")

        return ans
```

# Combinations

## Difficulty
Medium

## Description
Given two integers `n` and `k`, return all possible combinations of `k` numbers chosen from the range `[1, n]`.

The combinations can be returned in any order.  
Combinations are unordered, meaning `[1,2]` and `[2,1]` are considered the same.

## Examples

**Example 1**
- Input: `n = 4, k = 2`
- Output: `[[1,2],[1,3],[1,4],[2,3],[2,4],[3,4]]`
- Explanation: There are `4 choose 2 = 6` total combinations.

**Example 2**
- Input: `n = 1, k = 1`
- Output: `[[1]]`
- Explanation: There is `1 choose 1 = 1` total combination.

## Constraints
- `1 ≤ n ≤ 20`
- `1 ≤ k ≤ n`


```python
class Solution:
    def combine(self, n: int, k: int) -> List[List[int]]:
        ans = []


        def comb(x,n,k,ans,curr):
            if k == 0:
                ans.append(curr)
                return
            if x>n:
                return
            comb(x+1,n,k-1,ans,curr+[x])
            comb(x+1,n,k,ans,curr)

        comb(1,n,k,ans,[])
        return ans
```

# Combination Sum

## Difficulty
Medium

## Description
Given an array of **distinct integers** `candidates` and a target integer `target`, return all **unique combinations** of candidates where the chosen numbers sum to `target`.

Rules:
- Each number in `candidates` may be used **unlimited times**.
- Two combinations are considered unique if the frequency of at least one chosen number is different.
- The combinations can be returned in any order.

## Examples

**Example 1**
- Input: `candidates = [2,3,6,7], target = 7`
- Output: `[[2,2,3],[7]]`
- Explanation:
  - `2 + 2 + 3 = 7`
  - `7 = 7`

**Example 2**
- Input: `candidates = [2,3,5], target = 8`
- Output: `[[2,2,2,2],[2,3,3],[3,5]]`

**Example 3**
- Input: `candidates = [2], target = 1`
- Output: `[]`

## Constraints
- `1 ≤ candidates.length ≤ 30`
- `2 ≤ candidates[i] ≤ 40`
- All elements of `candidates` are distinct
- `1 ≤ target ≤ 40`


```python
class Solution:
    def combinationSum(self, candidates: List[int], target: int) -> List[List[int]]:
        ans = []
        n = len(candidates) - 1

        def combSum(start,target,candidates,curr,n,ans):
            if target == 0:
                ans.append(curr)
                return

            if start>n:
                return
            if target<0:
                return

            combSum(start, target -candidates[start],candidates,curr+[candidates [start]],n,ans)
            combSum(start+1,target, candidates,curr,n,ans)

        combSum(0, target, candidates,[],n,ans)

        return ans
            
        
```
